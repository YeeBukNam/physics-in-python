```python
import numpy as np
import matplotlib.pyplot as plt


def F(n, X):
    """
                   1
        F (t) = ------- sin ( (2n+1) t ) 
         n      2n + 1
    """
    return np.sin((2*n+1)*X)/(2*n+1)


def RectWave(n, X):
    """
          n
        Sigma  F (t)
         k=1
    """
    y = np.zeros(X.shape)
    for k in range(n+1):
        y = F(k, X) + y
    return y


X = np.linspace(-5, 5, 1000)

plt.plot(X, RectWave(1, X))
plt.show()
```
