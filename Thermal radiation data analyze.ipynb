```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

data = {
    'Variable1': [0.5, 0.6, 0.7, 0.8],
    'Variable2': [1661859375
, 1781601216
, 1902503711
, 2024574336]
}

df = pd.DataFrame(data)

slope, intercept, r_value, p_value, std_err = stats.linregress(df['Variable1'], df['Variable2'])

regression_line = slope * df['Variable1'] + intercept

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Variable1', y='Variable2', s=100, color='blue', label='Data Points')
plt.plot(df['Variable1'], regression_line, color='red', label='Regression Line')

plt.xlabel('radiation measured at the sensor (mV)', fontsize=14)
plt.ylabel('deltaT^4 (10^9K)', fontsize=14)
plt.title('stefan boltzmann law at low temperature', fontsize=16)
plt.legend()

print(f'상관계수 (Correlation coefficient): {r_value:.3f}')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

data = {
    'Variable1': [0.25, 6.94*10**(-3), 2.07*10**(-3), 9.77*10**(-4), 5.67*10**(-4)],
    'Variable2': [41.8, 2.4, 0.9, 0.4, 0.3]
}

df = pd.DataFrame(data)

slope, intercept, r_value, p_value, std_err = stats.linregress(df['Variable1'], df['Variable2'])

regression_line = slope * df['Variable1'] + intercept

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Variable1', y='Variable2', s=100, color='blue', label='Data Points')
plt.plot(df['Variable1'], regression_line, color='red', label='Regression Line')

plt.xlabel('distance(x^-2)', fontsize=14)
plt.ylabel('radiation(mV)', fontsize=14)
plt.title('y versus x^-2', fontsize=16)
plt.legend()

print(f'상관계수: {r_value:.3f}')

plt.show()
```
