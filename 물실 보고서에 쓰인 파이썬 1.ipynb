```python
import numpy as np
import matplotlib.pyplot as plt

# Define rectangular wave function
def rect_wave(t, T):
    w = 2*np.pi/T
    n_terms = 50
    f = 0
    for n in range(1, n_terms+1, 2):
        f += 4/(n*np.pi)*np.sin(n*w*t)
    return f

# Define derivative of rectangular wave function
def rect_wave_deriv(t, T):
    w = 2*np.pi/T
    n_terms = 100
    f = 0
    for n in range(1, n_terms+1, 2):
        f += 4/(n*np.pi)*n*w*np.cos(n*w*t)
    return f

# Set up time array
t = np.linspace(0, 4*np.pi, 1000)

# Calculate rectangular wave and its derivative
s = rect_wave(t, T=2*np.pi)
s_deriv = rect_wave_deriv(t, T=2*np.pi)

# Plot the results
fig, ax = plt.subplots(nrows=2, sharex=True, figsize=(8,6))
ax[0].plot(t, s)
ax[0].set_title('Rectangular Wave')
ax[0].set_ylabel('Amplitude')
ax[1].plot(t, s_deriv)
ax[1].set_title('Derivative of Rectangular Wave')
ax[1].set_ylabel('Amplitude')
ax[1].set_xlabel('Time (s)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Generate a triangular wave

t = np.linspace(0, 1, 500)

tri_wave = signal.sawtooth(2 * np.pi * 5 * t)


# Calculate the derivative of the triangular wave
tri_wave_deriv = np.gradient(tri_wave)

# Plot the triangular wave and its derivative
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
ax[0].plot(t, tri_wave)
ax[0].set_title('Triangular Wave')
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(t, tri_wave_deriv)
ax[1].set_title('Derivative of Triangular Wave')
ax[1].set_xlabel('Time')
ax[1].set_ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
