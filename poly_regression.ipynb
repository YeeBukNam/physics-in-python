```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


x_data = [2,12,22,32,42]
y_data = [41.8,2.4,0.9,0.4,0.3]

data = pd.DataFrame({'x': x_data, 'y': y_data})

def poly_regression(x, a, b):
    return a / x**2 + b

popt, pcov = curve_fit(poly_regression, data['x'], data['y'])

a, b = popt

plt.scatter(data['x'], data['y'], label='Data')
plt.plot(data['x'], poly_regression(data['x'], a, b), 'r-', label='Fit: a=%5.3f, b=%5.3f' % tuple(popt))
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
